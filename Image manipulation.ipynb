```python
#Image sharpen
from PIL import Image
from PIL import ImageFilter 
import matplotlib.pyplot as plt

#load the image
my_image=Image.open('2.jpg')
#Use sharpen functions
sharp=my_image.filter(ImageFilter.SHARPEN)
#Save the image
sharp.save('E:/yathin/image_sharpen.jpg')
sharp.show()
plt.imshow(sharp)
plt.show()
```

```python
#Image flip
import matplotlib.pyplot as plt
#Load theimage
img=Image.open('2.jpg')
plt.imshow(img)
plt.show()
#Use the flip function
flip=img.transpose(Image.FLIP_LEFT_RIGHT)

#SAVE THE IMAGE
flip.save('E:/yathin/image_flip.jpg')
plt.imshow(flip)
plt.show()
```
